# Columns Description

<b> Mpg:</b>           Miles per galon of fuel (continuous variable). <br>
<b> Cylinders:</b>     Number of engine cylinders (multi-valued discrete variable). <br>
<b> Displacement:</b>  Continuous variable.  <br>
<b> Horsepower:</b>    The power produced by engine to move the car (continuous variable). <br>
<b> Weight:</b>        Car weight (continuous variable). <br>
<b> Acceleration:</b>  The acceleration of an engine can get per second (continuous variable). <br>
<b> Model year:</b>    Car released year from 1970 to 1982 (multi-valued discrete variable). <br>
<b> Origin:</b>        Car manufacturing places (1 -> USA, 2 -> Europe, 3 -> Asia) (multi-valued discrete variable). <br>
<b> Car name:</b>      Car model name (unique for each instance). <br>

# Two class variables

<b>Origin vs Cylinders</b>

```python
df1=pd.crosstab(df['cylinders'], df['origin'])
sns.heatmap(df1, annot=True, cmap='viridis', linecolor='r', linewidth=2, fmt='g')
```

<b> Insights </b> <br>
<b>1</b>. Japan only 4 or 3 or 6 cylinders. <br>
<b>2</b>  Usa only 4,8 or 6 cylinders.<br>
<b>3</b>  The usa origin at 103 percentile as per other origins
